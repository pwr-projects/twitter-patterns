```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import decomposition
from sklearn import preprocessing
import seaborn as sns
```

```python
path_to_ds = '../../dataset/final_dataset/twitter_patterns_with_graph.csv'
data = pd.read_csv(path_to_ds)
```

```python
points_df = data.drop(['tp_author', 'tp_group'], axis=1)
```

```python
points = points_df.values
points = preprocessing.normalize(points)
labels = data['tp_group'].values
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(points)
pca_points = pca.transform(points)
```

```python
pca_points_x, pca_points_y = zip(*pca_points)
```

```python
pca_dataframe = pd.DataFrame(list(zip(pca_points_x, pca_points_y, labels)), columns=['x','y', 'group'])
```

```python
sns.lmplot(x='x', y='y', data=pca_dataframe, fit_reg=False, hue='group', legend=False)
plt.legend(loc='upper left')
```
